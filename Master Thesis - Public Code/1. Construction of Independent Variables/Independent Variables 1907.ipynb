make functions to calculate euclidean distance to universities

```python
# Define the Euclidean distance function
def euclidean_distance(lat1, lon1, lat2, lon2):
    return np.sqrt((lat2 - lat1)**2 + (lon2 - lon1)**2)

# Apply the function to each county
def min_distance_to_uni(row, df):
    # Filter out cities with a university
    uni_cities = df[df["uni_dummy"] == 1]
    
    # If the current city has a university return 0 as the distance
    if row["uni_dummy"] == 1:
        return 0
    
    # Calculate distances from the current city to all universities
    distances = uni_cities.apply(
        lambda uni_row: euclidean_distance(
            row["latitud"], row["longitud"],
            uni_row["latitud"], uni_row["longitud"]
        ),
        axis=1
    )
    
    # Return the minimum distance
    return distances.min()
```

```python
# Create the new column
control_df["min_distance_uni"] = control_df.apply(min_distance_to_uni, df=control_df, axis=1)
```
